```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir_name = './part1_rnn_input_len_big_range/'
logs = os.listdir(dir_name)
data = []
for log in logs:
    if log[-4:] == '.csv':
        experiment = {
            'label' : log[:-4],
            'data' : pd.read_csv(dir_name + log)
        }
        data.append(experiment)
```

```python
_, axes = plt.subplots(1,1)
cutoff = 150
for xp in data:
    axes.plot(xp['data']['Step'][:cutoff], xp['data']['Value'][:cutoff], label=xp['label'])
    axes.set_xlabel('Examples (x 128)')
    axes.set_ylabel('Accuracy')
axes.legend()
plt.savefig('rnn_range.png')
```
